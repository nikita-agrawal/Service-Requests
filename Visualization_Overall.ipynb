```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
plt.style.use('ggplot')
```

```python
df1 = pd.DataFrame(data = 
                   {"Complaint Type": ["Noise Related","Illegal parking","Heat/hot water","Blocked driveway",\
                                      "Street condition","Unsanitary condition","Non-emergency police matter", \
                                      "Water system", "Street light condition","Missed collection (all materials)"],
                    "Percent of Total": [29.81, 9.50, 7.67, 5.53, 2.68, 2.37, 2.08, 2.02, 1.93, 1.80]})                     
```

```python
df1
```

```python
df1.sort_values(by="Percent of Total").plot(x="Complaint Type",kind = 'barh', color = 'b',figsize=(8,6))
plt.title('Top Complaint Categories (2019 - 2021)')
plt.xlabel('Percent of Total Complaints')
#plt.ylabel('Complaint Type')   
plt.rcParams.update({'font.size': 12})
plt.savefig('figure_images/Most_Popular_Complaints.png', format='png', dpi=1200, bbox_inches='tight') # use format='svg' or 'pdf' for vectorial pictures
```
