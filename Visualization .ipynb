sr_noise_data['Year_month_created']=pd.DatetimeIndex(sr_noise_data['Date_created']).to_period('M').to_timestamp()

#Total number of complaints created every day
denominator = sr_noise_data.groupby(['Year_month_created'])['Unique Key'].count()

#merge the two tables
sr_noise_data_den = pd.merge(sr_noise_data,denominator, on='Year_month_created')
sr_noise_data_den.rename(columns={'Unique Key_x': 'Unique Key','Unique Key_y': 'Month_Count_All'},inplace=True)

numerator = sr_noise_data_den.groupby(['Noise Descriptor',\
                                       'Year_month_created','Month_Count_All'])['Unique Key'].agg('count')

#merge again
sr_noise_data_den_num = pd.merge(sr_noise_data_den,\
                                 numerator,\
                                 how='left',\
                                 on=['Noise Descriptor','Year_month_created','Month_Count_All'])

#rename 
sr_noise_data_den_num.rename(columns={'Unique Key_x': 'Unique Key',\
                                      'Unique Key_y': 'Month_Count_Descriptor'},\
                             inplace=True)

sr_noise_data_den_num['Ratio_Descriptor_Monthly'] = \
                     sr_noise_data_den_num['Month_Count_Descriptor'] / sr_noise_data_den_num['Month_Count_All'] 

import plotly.express as px
df = sr_noise_data_den_num 
fig = px.line(df, x="Year_month_created", y="Ratio_Descriptor_Monthly", color="Noise Descriptor")
fig.show()

df = sr_noise_data_den_num 
fig = px.line(df, x="Year_month_created", y="Ratio_Descriptor_Monthly", color="Noise Descriptor",range_y=[0,.3])
fig.show()
